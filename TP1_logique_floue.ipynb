# TP1 : Implémentation de la logique floue avec scikit-fuzzy


## Partie 1 : Introduction théorique et installation
Dans cette partie, nous allons utiliser la bibliothèque `scikit-fuzzy` pour définir une variable floue pour la température. Nous allons aussi visualiser les fonctions d'appartenance pour les états linguistiques "Froid", "Moyen", et "Chaud".


```python

# Importer les bibliothèques nécessaires
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt

# Définir la plage de valeurs pour la température
x_temp = np.arange(0, 51, 1)

# Définir les fonctions d'appartenance pour Froid, Moyen, et Chaud
froid = fuzz.trapmf(x_temp, [0, 0, 10, 20])
moyen = fuzz.trapmf(x_temp, [15, 20, 30, 35])
chaud = fuzz.trapmf(x_temp, [30, 40, 50, 50])

# Visualiser les fonctions d'appartenance
plt.figure()
plt.plot(x_temp, froid, 'b', label='Froid')
plt.plot(x_temp, moyen, 'g', label='Moyen')
plt.plot(x_temp, chaud, 'r', label='Chaud')
plt.title('Fonctions d appartenance pour la Température')
plt.xlabel('Température (°C)')
plt.ylabel('Degré d appartenance')
plt.legend()
plt.show()
```


## Partie 2 : Système flou pour le contrôle de température
Nous allons créer un système flou qui ajuste la puissance d’un chauffage en fonction de la température ambiante. Les variables sont :
- **Entrée** : Température (Froid, Moyen, Chaud).
- **Sortie** : Puissance de chauffage (Faible, Moyenne, Forte).


```python

# Définir la plage de valeurs pour la puissance de chauffage
x_chauffage = np.arange(0, 101, 1)

# Définir les fonctions d'appartenance pour Faible, Moyenne, et Forte
faible = fuzz.trapmf(x_chauffage, [0, 0, 30, 50])
moyenne = fuzz.trapmf(x_chauffage, [30, 50, 70, 90])
forte = fuzz.trapmf(x_chauffage, [70, 90, 100, 100])

# Visualiser les fonctions d'appartenance de chauffage
plt.figure()
plt.plot(x_chauffage, faible, 'b', label='Faible')
plt.plot(x_chauffage, moyenne, 'g', label='Moyenne')
plt.plot(x_chauffage, forte, 'r', label='Forte')
plt.title('Fonctions d\'appartenance pour la Puissance de Chauffage')
plt.xlabel('Puissance (%)')
plt.ylabel('Degré d\'appartenance')
plt.legend()
plt.show()
```


## Partie 3 : Exploitation avancée de la bibliothèque scikit-fuzzy
Dans cette partie, nous concevons un système de gestion de la vitesse d’un véhicule en fonction de deux entrées : la distance de l’obstacle et la visibilité.


```python

# Plage de valeurs pour Distance et Visibilité
x_distance = np.arange(0, 101, 1)
x_visibilite = np.arange(0, 101, 1)

# Définir les fonctions d'appartenance pour Distance (Proche, Moyenne, Loin)
proche = fuzz.trapmf(x_distance, [0, 0, 30, 50])
moyenne_distance = fuzz.trapmf(x_distance, [30, 50, 70, 90])
loin = fuzz.trapmf(x_distance, [70, 90, 100, 100])

# Définir les fonctions d'appartenance pour Visibilité (Faible, Moyenne, Bonne)
faible_visibilite = fuzz.trapmf(x_visibilite, [0, 0, 30, 50])
moyenne_visibilite = fuzz.trapmf(x_visibilite, [30, 50, 70, 90])
bonne_visibilite = fuzz.trapmf(x_visibilite, [70, 90, 100, 100])

# Visualiser les fonctions d'appartenance de Distance et Visibilité
plt.figure(figsize=(10, 4))
plt.subplot(121)
plt.plot(x_distance, proche, 'b', label='Proche')
plt.plot(x_distance, moyenne_distance, 'g', label='Moyenne')
plt.plot(x_distance, loin, 'r', label='Loin')
plt.title('Fonctions d\'appartenance pour la Distance')
plt.xlabel('Distance')
plt.ylabel('Degré d''appartenance')
plt.legend()

plt.subplot(122)
plt.plot(x_visibilite, faible_visibilite, 'b', label='Faible')
plt.plot(x_visibilite, moyenne_visibilite, 'g', label='Moyenne')
plt.plot(x_visibilite, bonne_visibilite, 'r', label='Bonne')
plt.title('Fonctions d\'appartenance pour la Visibilité')
plt.xlabel('Visibilité')
plt.ylabel('Degré d\'appartenance')
plt.legend()
plt.show()
```
